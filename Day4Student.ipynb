```python
import requests
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Replace 'your_api_key' with your actual OpenWeatherMap API key
api_key = 'eaad329271adc36b53c4ec13b9070728'
city = 'Cluj-Napoca'  # You can change the city to any location
url = f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={api_key}&units=metric"

# Fetch weather data
response = requests.get(url)
weather_data = response.json()

# Extract temperature and humidity data for each time point
temperatures = []
humidities = []

for forecast in weather_data['list']:
    temperatures.append(forecast['main']['temp'])
    humidities.append(forecast['main']['humidity'])

# Convert lists to numpy arrays
temperatures = np.array(temperatures)
humidities = np.array(humidities)

# Calculate Pearson correlation
correlation, _ = pearsonr(temperatures, humidities)
print(f"Pearson correlation between temperature and humidity: {correlation}")

# Plot the correlation
plt.scatter(temperatures, humidities)
plt.title(f'Temperature vs. Humidity in {city}')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Humidity (%)')
plt.show()
```
